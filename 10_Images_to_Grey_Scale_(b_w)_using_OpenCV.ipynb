```python
# Input folder path
input_folder = './Coloured_Images/'

# Output folder path
output_folder = './Grayscale_Images/'

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Iterate over files in the input folder
for filename in os.listdir(input_folder):
    # Construct the full path for the input and output files
    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, f"Grayscale_{filename}")

    # Read the image
    image = cv2.imread(input_path)
    # Convert to grayscale
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Save the grayscale image
    cv2.imwrite(output_path, grayscale)
```
